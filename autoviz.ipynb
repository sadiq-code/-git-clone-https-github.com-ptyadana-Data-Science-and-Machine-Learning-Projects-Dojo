```python
from sklearn.datasets import load_iris
data = load_iris()
```

```python
print(data)
```

```python
full_data = np.append(data.data,data.target.reshape(-1,1),axis=1)
```

```python
print(full_data)
```

```python
df=pd.DataFrame(full_data)
df.columns=data.feature_names+["type"]
```

```python
print(df)
```

```python
df.to_csv("iris.csv")
```
